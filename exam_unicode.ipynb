```python
# 파이썬 유니코드 / 인코딩 핵심정리

# 파이썬에서 문자열은 컴퓨터 내부에서 숫자(유니코드)로 저장된다.
# - 모든 문자는 고유한 숫자(유니코드)로 저장
# - ex) 각 글자를 → 유니코드(Unicode)로 저장 

# 1.
text = '파이썬'

print(text) #문자열 출력
print(len(text)) # 문자열 길이

# 2.
# 문자 / 숫자 변환
# String: 문자열
# Chat: 문자
print(ord('가')) # 문자를 유니코드 숫자로 변환하겠다(ord:순서번호, 위치번호)
print(chr(44032)) # 숫자를 다시 문자로 변환



# 3.
# 문자열을 바이트로 변환 (encode)
# 1Byte 당 1bit 이다
# - encode: 문자열을 바이트(byte)로 바꾸는 작업이다.
# 파일 저장이나 네트워크 전송 시 사용됨

encoded = text.encode('utf-8') # utf-8 은 '한글'프로그램이라고 생각하면 쉬움 // 표준 인코딩

print(encoded)
print(type(encoded)) #어떤 타입인지 알아보는 방법



# 4.
# 바이트(byte) 를 다시 문자열로 변환 (decode)

decoded = encoded.decode('utf-8')

print(decoded)
print(type(decoded))



# 5.
# 인코딩이 맞지 않으면 오류가 발생함
# 그럴때 변경하는 방법
# 원래 인코딩/디코딩 할때 같은 타입으로 해줘야 하나 밑에 예시처럼 다르게 쓰면 오류발생!!
#===예제 시작
try: 
    wrong = encoded.decode('cp949')
    print(wrong)
# ===예제 끝

except UnicodeDecodeError: # 다시 맞는 타입코드로 변경해주는 방법
    print('인코딩 방식이 맞지 않아 오류 발생')


# unicode: 문자 → 번호
# UTF-8: 번호 → byte(바이트)
# encode: 문자열 → bytes
# decode: bytes → 문자열
```
