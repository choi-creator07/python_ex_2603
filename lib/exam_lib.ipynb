```python
# 데이터 형식 모듈
# JSON 형식으로 데이터를 변환
#   - JSON: 데이터를 약속된 모양으로 적는 방식("Key", "Value" 로 구분")
# 파이썬에서는 JSON을 다루기 쉬움
# JSON 특징: 웹, API 파일 저장에 사용
#   -API란?: 프로그램끼리 대화하는 방법(규칙)

import json
data = {
    "name" : "jin",
    "age" : 22
}

print(json.dumps(data))
```

```python
# 정규표현식 모듈
# 문자열 패턴 검사 모듈 // 문자열 치환 // 텍스트 파싱
# 형식 확인(아이디, 이메일 등)에 사용
# True/False로 출력됨


import re # 정규표현식 사용방법

# 문자열이 "hi"로 시작하는지 확인 하는 방법
print(bool(re.match('hi', 'hi python')))
```
