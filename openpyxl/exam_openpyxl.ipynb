```python
from openpyxl import * 

# ===딕셔너리로 값(여러개) 생성 후 엑셀에 넣기===

item1 = {
    '품목': '바나나',
    '가격': '3000',
    '수량': 1,
    '날짜': '2026년 3월 10일'
}

item2 = {
    '품목': '책',
    '가격': '15000',
    '수량': 10,
    '날짜': '2026년 3월 11일'
}

item3 = {
    '품목': '커피',
    '가격': '3500',
    '수량': 20,
    '날짜': '2026년 3월 12일'
}

item4 = {
    '품목': '신발',
    '가격': '50000',
    '수량': 4,
    '날짜': '2026년 3월 13일'
}


wb = Workbook() #새로운 워크북을 만들어 저장
ws = wb.active

ws['A1'] = '품목' # A열 1행: 품목
ws['B1'] = '가격' # B열 1행: 가격
ws['C1'] = '수량' # C열 1행: 수량
ws['D1'] = '날짜' # D열 1행: 날짜



items = [item1, item2, item3, item4]






# 반복문을 돌려서 넣어보기!!

i = 2 #2부터 시작하는 이유는 위에가 1열이기 때문!!
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['A' + str(i)] = item['가격']
    ws['A' + str(i)] = item['수량']
    ws['A' + str(i)] = item['날짜']
    i += 1

wb.save('test.xlsx') # 변경 값 저장
```

```python
from openpyxl import * 
from openpyxl.styles import PatternFill # 셀 배경색을 바꾸기 위한 것을 불러오기

#===셀 배경 색 바꾸기===
wb =Workbook()
ws = wb.active # 워크시트 열기


# 헤더
ws['A1'] = '이름'
ws['B1'] = '국어'
ws['C1'] = '영어'
ws['D1'] = '수학'

color = PatternFill(fill_type='solid', fgColor='51c3ed') # 'solid'는 직선! 이라는 뜻 //'fgColor'는 [color picker]를 검색하여 원하는 색상을 작성해도 됨





# 반복문을 통해 셀의 배경색 바꾸기
for cell in ws[1]: # 1행 전체
    cell.fill = color

wb.save('bg_color.xlsx') # 변경한 색상 저장하는 방법

```

```python
from openpyxl import *
from openpyxl.chart import BarChart, Reference


# === 차트 만들기 ===

wb = Workbook()
ws = wb.active


# 예제 데이터 입력
ws.append(['Name', 'Score']) # 헤더 행
ws.append(['홍길동', '90'])
ws.append(['임꺽정', '80'])
ws.append(['홍길순', '70'])
ws.append(['임꺽순', '92'])


# BarChart 객체 생성(막대그래프)
chart = BarChart()

# 차트에 사용할 데이터 범위 지정
# ws → 차트이름
# min_col=2 → B열(Scrore)
# min_row=1 → 1행부터
# max_row=5 → 5행까지
data = Reference(ws, min_col=2, min_row=1, max_row=5)


# 지정한 데이터를 차트에 추가
# 첫 행을 데이터 제목(Series name)으로 사용한다는 의미임
chart.add_data(data, titles_from_data=True)


#차트 제목 설정
chart.title = "Student Scores"

# 차트를 워크시트의 E5 위치부터 추가하겠다는 의미
ws.add_chart(chart, "E5")



wb.save('chart_exam.xlsx')
```
