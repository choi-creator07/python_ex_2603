```python
# scrap @@@ 가격, 리뷰 등 웹사이트의 데이터 정보를 추출하는 방법 (모듈 설치 필요)
# requests : 웹 서버에 요청(http)을 보내고 응답(html)을 받는 것

import requests as req

from bs4 import BeautifulSoup # BeautifulSoup 은

# 요청시 오류 났을 때 아래 문구 입력
# User-agent: 웹 요청을 보낼때 프로그램이 아니라 브라우저 처럼 요청하는것
#  --------여기부터
request_headers = {'User-Agent': ('Mozilla/5.0 (Windows NT 10.0;Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98Safari/537.36'), }
# 여기까지 --------- 꼭 코드 붙여서 쓰기!!!!!! @@@@@@@@@@중요@@@@@@@@@@----------

response = req.get('https://www.naver.com', request_headers)   # response: 응답 받겠다
print(response)

# 예제 html
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""


# print(html_doc)

soup = BeautifulSoup(html_doc, 'html.parser')
# soup 을 이용해서 html_doc 문자열 파싱(원하는 문구만 출력), 그 내용을 soup 변수에 대입

a_list = soup.select('a')

print("== a태그들 리스트 ==")
print(a_list)


# 선택하는 법
print("== a태그 리스트 중 첫번째 ==")
print(a_list[0])


# 내용만 가져오고 싶다
print("== a태그 리스트 중 첫번째 텍스트 ==")
print(a_list[0].text)


# 링크를 가져오고 싶다(속성선택)//대괄호[] 사용
print("== a태그 리스트 중 첫번째 링크 ==")
print(a_list[0]['href']) # 속성을 선택할 땐 대괄호 ///규칙: 이름[순서]['리스트명']

# 리스트에 있는 텍스트 모두 출력
print("== a태그 리스트에 있는 텍스트 모두출력==")
for a in a_list:
    print(a.text)

# 리스트에 있는 링크 모두 출력
print("== a태그 리스트에 있는 링크 모두출력==")
for link in a_list:
    print(link['href'])

```
