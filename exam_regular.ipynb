```python
# regular expression :정규 표현식 / 글자 또는 숫자등을 찾기

# 정규표현식 실습 전 글자찾기 예시
text = '안녕하세요. 저는 15살 입니다.'
print("15" in text) # True / False
```

```python
# 파이썬에서 정규표현식을 사용하려 re 모듈을 불러와야함@@@@@@@@@@@@@@@@@@@@

import re #모듈불러오기

text = '안녕하세요. 저는 15살 입니다. 저는 안경이 15개 있습니다.' # 15라는 숫자가 2개 들어있음


# r : row string \ (특수문자) // 역슬래시 '\'는 특별함
# \d :숫자를 의미 (서식지정)
# + : 1개 이상

Pattern = r"\d+" # 찾고싶은 규칙!

result = re.findall(Pattern, text) # findall(찾을 규칙, 대상 문자열) = 문자열에서 패턴과 맞는 내용을 전부 찾아서 리스트로 반환
print(result)
```

```python
# 정규표현식으로 특정 문자열을 찾는 방법
# 패턴찾기

sample = 'abc123xyz456'

print(re.findall(r'\d+', sample)) # 대상을 1개 이상 찾아라
print(re.findall(r'[a-z]+', sample)) # 소문자 [a-z]까지 이어진 것을 찾아라
```

```python
# 특정 형식 찾기

data = 'test123@gmail.com'

# \w : 문자 또는 숫자
# + : 1개이상
# @ : 말그대로 @
# w. : 점(특수문자이기 때문에'\'를 붙임)
email_pattern = r'\w+@\w+\.\w+' # 이메일 찾는 규칙은 복잡하니 주의!! * [형식]: 문자/숫자 + @ + 문자/숫지 + . + 문자/숫자(도메인)

emails = re.findall(email_pattern, data)
print(emails)
```

```python
# 문자열 치환

test = "가격은 1200원 입니다."

# sub('찾을 패턴', '바꿀문자', 대상 문자열)
# 정규표현식의 문자열에서 특정패턴을 찾아서 문자로 바꾸는 함수
new_text = re.sub(r'\d+', '', test) # 숫자를 찾아서 빈문자로 바꾸겠다
print(new_text)
```

```python
# 정규표현식 vs 일반 문자열 처리

test = '몸무게" 70kg, 키: 170cm'

test.split(',') #split 은 문자열 쪼개기 // 단, 이중에서 숫자만 고를 순 없으나 아래 코드처럼 가능
nums = re.findall(r'\d+', test)
print(nums)
```
