```python
# 인스턴스 속성, 기능


class Person():
    def __init__(self): # __init__ 메서드는 세팅값
        self.hello = '안녕하세요' # 인스턴스 변수


    def hi(self):
        print(self.hello) # 클래스 설계도

jin = Person() # 이때 __init__ 메서드는 실행됐다
# jin = hi()
```

```python
class Person():
    def __init__(self, name, age, hobby):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby

    def introduce(self):
        print(f"{ self.hello}, 저는 {self.name} {self.age}살 입니다." ) # 클래스 안에서는 객체를 부를때 self로 하지만 외부에서는 jin으로 부른다!! 주의!!!!

jin = Person('진', 22, '운동')
jin.introduce()


print(f"나이 : {jin.age}") #클래스 외부에서는 jin으로 부른다!!
print(f"이름 : {jin.name}")
print(f"취미 : {jin.hobby}")
```

```python
#z클래스 외부에서는 3접근 불가
#


class Person():
    def __init__(self, name, age, hobby, weight):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
        self.weight = weight
    def diet(self, loss_weight): # 비공개 속성
        self._weight -= loss_weight
        print(self.weight)

jin = Person('진', 22, '운동', 88)
jin.diet(10)

print(jin.weight)
```

```python
# 강사꺼


# 클래스 외부에서는 접근 불가, 
# 클래스 내부에서만 사용 가능한 private attribute

class Person():
    def __init__(self, name, age, hobby, weight):
        self.hello = '안녕하세요'
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight # 비공개 속성
    
    def diet(self, loss_weight):
        self.__weight -= loss_weight
        print(self.__weight)

jin = Person('진', 22, '운동', 88)
jin.diet(10)
```
