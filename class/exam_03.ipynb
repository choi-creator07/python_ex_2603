```python
# 클래스 속성 만들기

class Person:
    def __init__(self):
        self.hello = "안녕하세요" # 인스턴스 속성

class Person:
    folder = [] # 클래스 속성

    def add_files(self, file):
        self.folder.append(file)


jin = Person()
jin.add_files('txt')


print(jin.folder)
```

```python
# 클래스 속성 만들기_2

class Person:
    folder = [] # 클래스 속성 (클래스 속성은 개별로 만드는게 아니라 클래스 속성은 공공재 이다.) // 모든 인스턴스에서 접근이 가능함

    def add_files(self, file):
        self.folder.append(file)


jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')


print(jin.folder)
print(paul.folder)
```

```python
# 인스턴스 속성

class Person:
    def __init__(self):
        self.folder = []
    
    def add_files(self, file):
        self.folder.append(file)

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')


print(jin.folder)
print(paul.folder)
```

```python
# 비공개 클래스 속성 만들기


class Person:
    __weight = 78  # 비공개 인스턴스를 만들때는 앞에 언더바(_) 2개 입력

    def print_weight(self):
        print(Person.__weight)

jin = Person()
jin.print_weight()

# print(jin.__weight)





# 인스턴스 속성: 각각의 객체가 가지는 속성이다.
# 클래스 속성: 속성이 클래승 딱 하나 존재하며 모든 인스턴스가 공유하는 속성이다.
```
