```python
# 클래스란?
# → 객체를 표현하기 위한 문법의 일종, 프로그래밍으로 객체를 만들 때 사용하는 것
# Python은 객체지향언어!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# 객체: object (표현가능) → 현실, 가상, 개념 등

# 속성(명사적 특징) → attribute _객체의 속성
# 능력/기능(동사적 특징) → method _클래스 안에서의 함수

```

```python
# 클래스(=설계도) 만드는 방법 (class명은 항상 대문자로 시작!!)
# 객체가 객체를 지칭할때 object
# 객체가 클래스와 연관지어서 지칭할때 instance


class Person:
    def Hello(self): # 메서드 (def 부터는 메서드이다. 클래스에 종속됨)
        print("hi")

    def hi(self): # 
        self.Hello()

jin = Person() # jin은 Person 의 인스턴스(변수/객체)이다.
jin.Hello() #인스턴스 메서드 (메서드 호출방법 // 메서드 호출은 해당 인스턴스로 호출해야한다)

jin.hi()



```
