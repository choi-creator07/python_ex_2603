```python
#요소찾는방법 3가지 / 활용방법
#   - 태그이름(TAG NAME)
#   - class / id(CLASS_NAME / ID)
#   - xPath
```

```python
#요소찾는방법 3가지


from selenium import webdriver
from selenium.webdriver.common.by import By

# ==========================================================
# 0) 브라우저 실행 + 페이지 접속
# ==========================================================

driver = webdriver.Chrome()
driver.get('https://quotes.toscrape.com/') # 괄호안의 사이트를 실행하겠다


# ==========================================================
# 1) 태그이름(TAG_NAME)으로 요소찾기
# ==========================================================
''' 여러줄 주석
- HTML 태그(div, span, a 등)으로 요소찾기
- 같은 태그가 많고 범위가 너무 넓어지는 경우가 많아 실무에서는 주로 '연습용'으로 사용
'''

div_elements = driver.find_elements(By.TAG_NAME, 'div') # div_elements는 div_element를 감싸주는 것 //'div'로 감싸진 태그찾기
print('TAG_NAME으로 찾은 div 태그 개수:', div_elements)
# 또는
print('TAG_NAME으로 찾은 div 태그 개수:', len(div_elements)) # 보기가 어려울 수 있기 때문에 개수사용 추천



# ==========================================================
# 2) Class로 요소찾기
# ==========================================================
'''
- HTML에서 class='text'처럼 지정된 class이름으로 요소찾기
Selenium에서 가장 자주 사용하는 방식 중 하나

★ 주의) class에 공백이 포함된경우(ex: class="a b")는 CLASS_NAME으로 바로 찾기가 어려움 //이유: 공백으로 이름을 구분하기 때문!!!!!!

'''
first_quote_text = driver.find_element(By.CLASS_NAME, 'text')
print('/\nCLASS_NAME으로 찾은 첫번째 명언: ', first_quote_text.text)


# ==========================================================
# 3) id로 요소찾기(ID)
# ==========================================================
'''
- id 는 페이지내에 '유일'하게 존재하는 것이다.
- 가장 정확하게 찾고 싶을때 가장 안정적임

'''
# 예시
#quote_container = driver.find_element(By.ID, 'quote')
#print("\nID로 찾은 요소의 태그이름:", quote_container.tag_name) # 얘는 현재 해당 사이트에 없어서 오류남

# ==========================================================
# 4) find_element vs find_elements
# ==========================================================
'''
- find_element: 요소 1개만 찾음 (요소가 없으면 에러 발생)
- find_elements: 요소 여러개를 찾고 리스트로 반환 (요소가 없으면 빈 리스트)
'''
one_quote_block = driver.find_element(By.CLASS_NAME,'quote')
many_quote_block = driver.find_elements(By.CLASS_NAME,'quote')
print("\nfind_element 로 찾은 quote 블록 1개: ", type(one_quote_block))
print("find_elements 로 찾은 quote 블록 개수: ", len(many_quote_block))

# ==========================================================
# 5) xPath
# ==========================================================
'''
- xPath는 HTML 의 구조(경로)를 이용해서 요소를 찾는 방법
- id나 class로 찾기 어려울 때 사용
- HTML 트리구조를 따라가면서 원하는 요소를 지정할 수 있음

<장점>
- 조건이나 경로로 정확하게 요소 선택 가능

<단점>
- 페이지 구조가 바뀌면 xPath도 같이 깨질 수 있음
'''

# 첫 번때 명언 텍스트를 xPath로 찾기

xPath_quote_text = driver.find_element(
    By.XPATH,
    "//div[@class='quote']/span[@class='text']" #XPATH 가 안쪽 자식/후손 선택자를 찾는 것이다. // 경로 기반 선택
)

'''
[위의 코드 설명]
- // : 문서 전체에서 검색(:root부터 전역 탐색)
- div[@class='quote']: dic 태그 중 class='quote'인 오소
- / : 바로 아래 자식(child)관계
- span[@class='text']: 위에서 찾은 div 바로 아래에 있는 span 중 class='text' 인 요소

<의미>
- class='quote'인 div 아래(밑에) 자식 중 class='text' 인 span을 찾아라

'''

print('\nxPath로 찾은 첫 번째 명언: ', xPath_quote_text.text)


# 첫 번째 quote블록만 특정해서 찾는 xPath예시
xPath_first_block = driver.find_element(By.XPATH, "(//div[@class='quote'])[1]") #xPath를 사용하여 첫 번째 인거 찾기 xPath는 index가 [0]이 아닌 [1]부터 시작
xPath_first_author = xPath_first_block.find_element(By.CLASS_NAME, "author") # 저자 찾기

print('xPath로 찾은 첫번째 저자: ', xPath_first_author.text)


# 브라우저 종료
driver.quit() # 꼭 하기!!!!!!!
```
